## Introduction

Saudi Central Bank (SAMA) offers short-term securities that mature in one year or less called Treasury Bill (T-Bills). Unlike Treasury bonds which has long-term maturity, T-Bills are purchased for a price less than their face value (called discount rate) and then when they mature, the treasury (SAMA) pays the investor the full face value.

This notebook aims to predict treasury bill yields. The data has been obtained from SAMA official website which we had to scrape to perform our modeling.

A big thank you to [General Authority for Statistics](https://www.stats.gov.sa/) for providing me the required data in a timely manner. Their level of professionalism is something formidable.

## Modeling
<a id='modeling'></a>
<a href="#">Back to top</a>


### Linear Regression model

```python
# define X, y
X = df.iloc[:,1:-1]
y = df['TARGET']
```

```python
# split data
x_train, x_val, y_train, y_val = train_test_split(X,y, random_state=7)
```

```python
# define linear regression model
lm = LinearRegression()
lm.fit(x_train, y_train)

# predict on train
train_pred = lm.predict(x_train)
```

```python
lm.intercept_
```

```python
lm.coef_
```

> * An increase in repo rate, increases treasury bill returns
> * An increase in inflation rate, decreases treasury bill returns
> * An increase in money supply, decreases treasury bill returns

```python
# visualize train predictions
sns.jointplot(x=train_pred, y=y_train, kind='reg')
```

```python
train_score = lm.score(x_train,y_train)
print(f'R2 on training set: {train_score}')
rmse = mean_squared_error(y_train, train_pred, squared=False)
print(f'RMSE on training set: {rmse}')
mae = mean_absolute_error(y_train, train_pred)
print(f'MAE on training set: {mae}')
```

> tend to miss by 43 points

```python
# predict on validation
val_pred = lm.predict(x_val)
```

```python
# visualize val predictions
sns.jointplot(x=val_pred,y=y_val, kind='reg')
```

```python
val_score = lm.score(x_val, y_val)
print(f'R2 on validation set: {val_score}')
rmse = mean_squared_error(y_val, val_pred, squared=False)
print(f'RMSE on validation set: {rmse}')
mae = mean_absolute_error(y_val, val_pred)
print(f'MAE on validation set: {mae}')
```

> Our model tends to miss the prediction by almost 0.44 points

### Ridge linear regression model

```python
# scale data for ridge model
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge
ss = StandardScaler()
x_train_scaled = ss.fit_transform(x_train)
x_val_scaled = ss.transform(x_val)
```

```python
lm_reg = Ridge()

lm_reg.fit(x_train_scaled, y_train)
train_pred = lm_reg.predict(x_train_scaled)
```

```python
lm_reg.score(x_train_scaled, y_train)
train_score = lm_reg.score(x_train_scaled, y_train)
print(f'R2 on training set: {train_score}')
rmse = mean_squared_error(y_train, train_pred, squared=False)
print(f'RMSE on training set: {rmse}')
mae = mean_absolute_error(y_train, train_pred)
print(f'MAE on training set: {mae}')
```

```python
val_pred = lm_reg.predict(x_val_scaled)
val_score = lm_reg.score(x_val_scaled, y_val)
print(f'R2 on validation set: {val_score}')
rmse = mean_squared_error(y_val, val_pred, squared=False)
print(f'RMSE on validation set: {rmse}')
mae = mean_absolute_error(y_val, val_pred)
print(f'MAE on validation set: {mae}')
```

### Polynomial regression model

```python
# transform features
from sklearn.preprocessing import PolynomialFeatures

# define a feature poly model with a degree of 2
poly = PolynomialFeatures(degree=2)

# feature transforms
x_train_poly = poly.fit_transform(x_train)
x_val_poly = poly.transform(x_val)
```

```python
# define lm_poly model
lm_poly = LinearRegression()
lm_poly.fit(x_train_poly, y_train)
train_pred = lm_poly.predict(x_train_poly)
```

```python
lm_poly.coef_
```

```python
lm_poly.score(x_train_poly, y_train)
train_score = lm_poly.score(x_train_poly, y_train)
print(f'R2 on training set: {train_score}')
rmse = mean_squared_error(y_train, train_pred, squared=False)
print(f'RMSE on training set: {rmse}')
mae = mean_absolute_error(y_train, train_pred)
print(f'MAE on training set: {mae}')
```

```python
val_pred = lm_poly.predict(x_val_poly)
val_score = lm_poly.score(x_val_poly, y_val)
print(f'R2 on validation set: {val_score}')
rmse = mean_squared_error(y_val, val_pred, squared=False)
print(f'RMSE on validation set: {rmse}')
mae = mean_absolute_error(y_val, val_pred)
print(f'MAE on validation set: {mae}')
```

### Polynomial Regression


```python
cross_val_score(lm_poly, X, y, cv=kf, scoring='r2')
```

```python
# Code by Pranav Kaushik from Medium.com
def create_polynomial_regression_model(degree):
    poly_features = PolynomialFeatures(degree=degree)
    X_poly = poly_features.fit_transform(X)
    poly = LinearRegression()
    score = cross_val_score(poly, X_poly, y, cv=kf)
    return score
                            
poly_cv = []
degree = []
for i in range(1,4):
    poly_cv.append(create_polynomial_regression_model(i))
    degree.append(i)
# plt.scatter(range(1,4),poly_cv)
for k in range(len(poly_cv)):
    print(f"Mean of degree {degree[k]}: {np.mean(poly_cv[k])}, Standard deviation: {np.std(poly_cv[k])}")
```
